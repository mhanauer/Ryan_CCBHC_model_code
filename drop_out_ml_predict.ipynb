```python
prob_drop =   pd.DataFrame(pred_prob[:,1], columns = ["prob_drop"])
def if_else(row):

    if row['prob_drop'] > 0.3212862:

        val = "very high risk"

    elif row['prob_drop'] > 0.3212862 / 2:

        val = "high risk "
    
    elif row['prob_drop'] >0.3212862 / 3:
        
        val = "medium risk"
    
    else:

        val = "low risk"

    return val

prob_drop['risk_level'] = prob_drop.apply(if_else, axis=1)
prob_drop
```
